# Data Visualization Tutorial with Python

Import pyplot from Matplotlib Library which is the library used for ploting graphs.

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("dataset.csv")
```

```python
data.corr()
```

```python
x = data["Healthy life expectancy at birth"]
y = data["Log GDP per capita"]
```

```python
x = x.sort_values()
y = y.sort_values()
plt.xlabel("Healthy life expectancy at birth")
plt.ylabel("Log GDP per capita")
plt.title("Relationship between Healthy life expectancy at birth and Log GDP per capita")
rects1 = plt.plot(x, y)
```

The increase in country's GDP per Capita will gradually increase the helth expectancy of a new born
