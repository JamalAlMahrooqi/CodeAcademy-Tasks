```python
x=data["Temperature (°C)"]
y=data["pH"]


plt.xlabel("Temperature (°C)")
plt.ylabel("pH")
plt.title("Relationship between Temperature and PH level of water")
area=x*10
plt.scatter(x,y,s=area,alpha=0.5)
```

```python
x=data["Conductivity (µS/cm)"]
y=data["pH"]
plt.xlabel("Conductivity (µS/cm)")
plt.ylabel("pH")
plt.title("Relationship between Conductivity and PH level of water")
area=x*0.3
plt.scatter(x,y,s=area,alpha=0.5)
```

pH level cuases the increase in Conductivity

```python
x=data["Dissolved Oxygen (mg/L)"]
y=data["pH"]
plt.xlabel("Dissolved Oxygen (mg/L)")
plt.ylabel("pH")
plt.title("Relationship between oxygen content and PH level of water")
area-x*2.5
plt.scatter(x,y,s=area,alpha=0.5)
```

As the pH level of water increases oxygen conttent in the water increases

```python
data2=pd.read_csv("Obesity Classification.csv")
data2
```

```python
count_male = data2['Gender'].value_counts()['Male']
count_Female = data2['Gender'].value_counts()['Female']
genders=('Male','Female')
val=(count_male,count_Female)
plt.bar(genders,val)
```

```python
Female_data = data2[data2['Gender'] == 'Female']
Male_data = data2[data2['Gender'] == 'Male']
#Female_data
```

```python
label=['Underweight','Normal Weight','Overweight']
underweight_count=Female_data['Label'].value_counts()['Underweight']
overweight_count=Female_data['Label'].value_counts()['Overweight']
normalweight_count=Female_data['Label'].value_counts()['Normal Weight']

data=[underweight_count,overweight_count,normalweight_count]
plt.pie(data, labels = label,shadow=False,autopct="%.2f%%");
plt.title("Female")
```

```python
label=['Underweight','Normal Weight','Overweight']
underweight_count=Male_data['Label'].value_counts()['Underweight']
overweight_count=Male_data['Label'].value_counts()['Overweight']
normalweight_count=Male_data['Label'].value_counts()['Normal Weight']

data=[underweight_count,overweight_count,normalweight_count]
plt.pie(data, labels = label,shadow=False,autopct="%.2f%%");
plt.title("Male")
```
