```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
gold=pd.read_csv("gold-prices.csv")
movie=pd.read_csv("movie_dataset.csv")
oil=pd.read_csv("oil.csv")
```

```python
movie
```

```python
data=movie.corr()
sns.heatmap(data,annot=True, cmap="Blues")
```

```python
movie_data=movie[['budget','revenue']]
movie_data
```

```python
x=movie_data['budget']
y=movie_data['revenue']
plt.scatter(x,y)
```

```python
l_model=sns.lmplot(x='budget', y='revenue', data=movie,order=1,ci=80,height=7, aspect=2 ,
          scatter_kws={"marker":"o","color":"indianred"});
```

```python
sns.residplot(data=movie,x='budget',y='revenue',lowess=True, line_kws=dict(color="r"))
sns.set(rc={'figure.figsize':(110.70,100)})
```
