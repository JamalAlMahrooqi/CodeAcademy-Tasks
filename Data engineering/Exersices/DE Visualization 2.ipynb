```python
import numpy as np
from numpy.random import randn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=randn(100)
dataset2=randn(100)
```

```python
np.std(dataset)
np.std(dataset2)
```

```python
plt.hist(dataset,color='red',edgecolor='black');
```

Density Graph

```python
sns.distplot(dataset2);
```

```python
plt.hist(dataset,density=True,color='indianred',alpha=0.5,bins=20)
plt.hist(dataset2,density=True,alpha=0.5,bins=20);
```

```python
sns.jointplot(dataset,dataset2);
```

```python
jp=sns.jointplot(dataset,dataset2,kind='kde')
jp.set_axis_labels('Data','Dataset2')
```

```python
sns.boxplot(dataset)
```

```python
sns.kdeplot(dataset);
```
